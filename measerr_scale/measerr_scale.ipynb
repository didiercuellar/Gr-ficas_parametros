```python
import pandas as pd 
import numpy as np 
import scipy.stats as st 
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt 
import math
import statistics as stats
```

```python
data=pd.read_csv('measerr_scale_U.csv',header=0,)
data = data.dropna(how='any')#quita los valores nan del documento
y=data.iloc[:,0]
w=data.iloc[:,1]
x=data.iloc[:,2]

print(stats.stdev(x))
print(data)
```

```python
yerror = w
xerror = 0.40483859217506146
plt.plot(x,y,'ro')
plt.errorbar(x,y,yerr=yerror, fmt = ' ' , ecolor='black')
plt.errorbar(x,y,xerr=xerror, fmt = ' ' , ecolor='blue')
plt.ylabel('measerr_scale_U')
plt.xlabel('$Hβ [log(M BH,vir /M_\odot)]$')
plt.title("Masa_virial_U")
#plt.xscale('log')
#plt.yscale('log')
plt.grid()
plt.savefig("masa del virial _measerr_scale_U_error.pdf")
plt.show
```
